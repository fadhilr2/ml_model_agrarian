```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import  OneHotEncoder, LabelEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv("../../../data/crop_recommendation_model_ready_data.csv")
X = df.iloc[:, :-1].values
y = df.iloc[:, -1]
```

```python
print(X)
```

```python
encoder = LabelEncoder()
y = encoder.fit_transform(y)
```

```python
ct_X = ColumnTransformer(transformers=[
                                       ("sun_exposure", OrdinalEncoder(), [-1])], remainder="passthrough")
X = ct_X.fit_transform(X)
```

```python
print(X)
```

```python
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
print(X_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[3:8] = sc.fit_transform(X_train[3:8])
X_test[3:8] = sc.transform(X_test[3:8])
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
print(encoder.inverse_transform(classifier.predict(X_test)))
```

```python
y_actual = encoder.inverse_transform(y_test)
y_pred = encoder.inverse_transform(classifier.predict(X_test))
print(np.concat([y_actual.reshape(-1, 1), y_pred.reshape(-1, 1)], 1))
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_actual, y_pred)

print(f"Model Accuracy: {acc:.2%}")
```
