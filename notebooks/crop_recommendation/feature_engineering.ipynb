```python
import pandas as pd
from numpy.random import randint
```

```python
df = pd.read_csv("./target.csv")
soil_df = pd.read_csv("./soil_data/soil_data.csv")
```

```python
#HUMIDITY FEATURE ENGINEERING
for i, r in df.iterrows():
  plant_info = crop_info[r["plant"]]
  area = plant_info["key_areas"][randint(len(plant_info["key_areas"]))]["name"]
  growing_season_start = plant_info["growing_season"][0]
  growing_season_end = plant_info["growing_season"][1]
  year = r["year"]
  year = int(year)


  humidity_df = pd.read_csv(f"./humidity_data/{area}.csv")
  year_filter_df = humidity_df[humidity_df["YEAR"] == year]
  day_filter_df = year_filter_df.iloc[growing_season_start-1:growing_season_end]

  RH2M = day_filter_df["RH2M"]
  avg_rh2m = RH2M.sum() / len(RH2M)
  
  df.iloc[i, 2] = avg_rh2m


for i, r in df.iterrows():
  plant_info = crop_info[r["plant"]]
  area = plant_info["key_areas"][randint(len(plant_info["key_areas"]))]

  soil_texture = soil_df[soil_df["area"] == area["name"]]
  df.iloc[i, 5] = soil_texture["soil_texture"].values[0]
df = df.drop(columns=['year'])
df.to_csv("../../data/crop_recommendation_model_ready_data.csv", index=False)

  
```
