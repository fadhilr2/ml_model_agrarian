```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df = pd.read_csv("../model_ready_data.csv")
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
ct = ColumnTransformer(transformers=[("encoder", OneHotEncoder(), [0])], remainder="passthrough")
X = np.array(ct.fit_transform(X))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
regressor = LinearRegression()
poly = PolynomialFeatures(degree=3)
X_poly_features = poly.fit_transform(X_train)
regressor.fit(X_poly_features, y_train)
```

```python
test_input = np.array(ct.transform([[
  "Carrots and turnips", 25.83762295, 18.24245902, 22.04004098, 7.595163934, 2139.885, 3.097377049, 0.747459016, 8.437459016]]))
print(regressor.predict(poly.transform(test_input)))
```

```python
y_pred = regressor.predict(poly.transform(X_test))

np.concatenate([y_pred.reshape(-1, 1), y_test.reshape(-1, 1)], 1)
```

```python
print("r2 " + str(r2_score(y_test, y_pred)))
print("mse " + str(mean_squared_error(y_test, y_pred)))
```
