```python
import pandas as pd
from numpy.random import randint
```

```python
df = pd.read_csv("./target.csv")
```

```python

def data_filler(area, year, growing_season_start, growing_season_end, base_temp):
  env_df = pd.read_csv(f"./fetched_environment_data/{area}.csv")

  filtered_year_df = env_df[env_df["YEAR"] == year]
  filtered_day_df = filtered_year_df[growing_season_start-1: growing_season_end]

  T2M_MAX_DF = filtered_day_df.iloc[:, 2]
  if(T2M_MAX_DF.empty):
    print(year)
    print(area)
    print(env_df)
    
  T2M_MIN_DF = filtered_day_df.iloc[:, 3]
  T2M_DF = filtered_day_df.iloc[:, 2:4]

  PRECIPITATION_DF = filtered_day_df.iloc[:, 4]
  SOIL_WETNESS_DF = filtered_day_df.iloc[:, 5]

  PAR_DF = filtered_day_df.iloc[:, -1]

  AVG_T2M_MAX = T2M_MAX_DF.sum() / len(T2M_MAX_DF)
  AVG_T2M_MIN = T2M_MIN_DF.sum() / len(T2M_MIN_DF)
  AVG_T2M = (AVG_T2M_MAX + AVG_T2M_MIN) / 2

  DIURNAL_TEMP_RANGE = AVG_T2M_MAX - AVG_T2M_MIN

  SUM_GDD = 0
  for i, r in T2M_DF.iterrows():
    T2M_MAX = r["T2M_MAX"]
    T2M_MIN = r["T2M_MIN"]
    GDD = ((T2M_MAX + T2M_MIN) / 2) - base_temp
    SUM_GDD = SUM_GDD + GDD

  AVG_PRECIPITATION = PRECIPITATION_DF.sum() / len(PRECIPITATION_DF)
  AVG_SOIL_WETNESS = SOIL_WETNESS_DF.sum() / len(SOIL_WETNESS_DF)

  AVG_PAR = PAR_DF.sum() / len(PAR_DF)

  # 2 = AVERAGE_MIN_TEMPERATURE
  # 3 = AVERATGE_MAX_TEMPERATURE
  # 4 = AVERAGE_TEMPERATURE
  # 5 = DIURNAL_TEMP_AVERAGE
  # 6 = SUM_GDD
  # 7 = AVG_PRECIPITATION
  # 9 = AVG_PAR
  return {
    '2': float(AVG_T2M_MIN),
    '3': float(AVG_T2M_MAX),
    '4': float(AVG_T2M),
    '5': float(DIURNAL_TEMP_RANGE),
    '6': float(SUM_GDD),
    '7': float(AVG_PRECIPITATION),
    '8': float(AVG_PAR)
  }
```

```python
for i, r in df.iterrows():
  plant_info = crop_info[r["plant"]]
  growing_season_start = plant_info["growing_season"][0]
  growing_season_end = plant_info["growing_season"][1]
  base_temp = plant_info["base_temp"]
  area = plant_info["key_areas"][randint(len(plant_info["key_areas"]))]["name"]
  yield_sq_meters = r["yield"] / 10000
  year = r["year"]


  data_fills = data_filler(area, year, growing_season_start, growing_season_end, base_temp)

  for k in data_fills.keys():
    df.iloc[i, int(k)] = data_fills[k]
    df.iloc[i, -1] = yield_sq_meters
    
df = df.drop(columns=['year'])
df.to_csv("../../../data/yield_prediction_model_ready_data.csv", index=False)
```
